```python
#for Title names
names = []

for i in range(10):
    print('Scraping page', i+1)
    name = browser.find_elements(By.XPATH, "//span[@class='a-size-medium a-color-base a-text-normal']")
    for n in name:
        names.append(n.text)
    
    sleep(3)
```

Here Remove the corrupted data from names/titles

```python
# Create a list to store the result
updated_names=[]


# List of indices to remove
indices_to_remove = [5, 23]  # Indices to remove

# Number of elements per cycle
segment_length = 24

# Iterate over the original list in cycles of 24
for i in range(0, len(names), segment_length):
    cycle = names[i:i+segment_length]  # Get a cycle of 24 elements

    # Remove elements at specified indices
    for j, name in enumerate(cycle):
        if j not in indices_to_remove:
            updated_names.append(name)
```

```python

# Initialize a list to store human names
human_names = []

# Process each name in the list
for author_name in author_names:
    doc = nlp(author_name)
    for sentence in doc.sentences:
        if any(entity.type == "PERSON" for entity in sentence.ents):
            human_names.append(author_name)

# Join the filtered names back into a list
filtered_names = list(set(human_names))  # Remove duplicates

print(filtered_names)
```
